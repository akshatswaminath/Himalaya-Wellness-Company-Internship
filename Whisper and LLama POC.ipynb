````python
from langchain import HuggingFacePipeline
from transformers import AutoTokenizer
import transformers
import torch

model = "meta-llama/Llama-2-7b-chat-hf"

tokenizer = AutoTokenizer.from_pretrained(model)

pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=1000,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)
````

````python
llm = HuggingFacePipeline(pipeline = pipeline, model_kwargs = {'temperature':0.5})
````

Write a concise summary of the following chunk text that is passed in tripple back quotes.
              Also give me a positive or negetive sentiment for the passed text.
              ```{chunk}```
              BULLET POINT SUMMARY:

````python
from langchain import PromptTemplate,  LLMChain

template = """
              Write a concise summary of the following {chunk}
              Also give me a positive or negetive sentiment for the {chunk}.

              BULLET POINT SUMMARY:
           """

prompt = PromptTemplate(template=template, input_variables=["chunk"])

llm_chain = LLMChain(prompt=prompt, llm=llm)
````
